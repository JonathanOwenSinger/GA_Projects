```python
#term = 36 months
#application_type = Individual

lc= accept_df[['acc_now_delinq','annual_inc','application_type','bc_util','delinq_2yrs','dti','emp_length','emp_title','fico_range_low','funded_amnt','funded_amnt_inv','grade','home_ownership','id','il_util','installment','int_rate','issue_d','last_fico_range_low','last_pymnt_d','loan_amnt','loan_status','next_pymnt_d','open_acc','percent_bc_gt_75','policy_code','pub_rec','purpose','sub_grade','term','title','tot_coll_amt','verification_status']]


```

```python
lc.set_index('issue_d', inplace=True)
```

```python
lc = lc[(lc['application_type'] == 'Individual') & (lc['term'] == ' 36 months')]
```

```python
lc['year'] = lc.index.year
lc['month'] = lc.index.month
```

```python
lc.set_index('id', inplace=True)
```

```python
lc.sort_values(['year', 'month'], ascending=[False, False]).tail()
```

```python
lc = lc[(lc['year'] <= 2015) & (lc['month'] <= 8) & (lc['month'] <= 8)]
```

```python
def experience(row):
    if row.emp_length == '< 1 year' or row.emp_length == '1 year':
        return ('1 year or less')
    elif row.emp_length == '2 years' or row.emp_length == '3 years' or row.emp_length == '4 years' or row.emp_length == '5 years':
        return ('2 to 5 years')
    elif row.emp_length == '1 year' or row.emp_length == '2 years' or row.emp_length == '3 years' or row.emp_length == '4 years' or row.emp_length == '5 years':
        return ('6 to 9 years')
    else:
        return ('10+ years')

lc['emp_length'] = lc.apply(lambda row: experience(row),axis = 1)
```

```python
def delinquent(row):
    if row.acc_now_delinq >= 1:
        return (1)
    else:
        return (0)
lc['acc_now_delinq'] = lc.apply(lambda row: delinquent(row),axis = 1)
```

```python
lc.drop(columns = ['emp_title','il_util','next_pymnt_d','open_acc','policy_code','term','bc_util','percent_bc_gt_75','tot_coll_amt'],inplace=True)
```

```python
def loan(row):
    if row.loan_status == 'Fully Paid' or row.loan_status == 'Current' or row.loan_status == 'Does not meet the credit policy. Status:Fully Paid':
        return (1)
    else:
        return (0)
lc['loan_status'] = lc.apply(lambda row: loan(row),axis = 1)

#1 means it was paid
#0 means it was charged off
```
